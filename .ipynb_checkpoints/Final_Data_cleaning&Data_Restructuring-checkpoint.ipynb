```python
def NLP(text):
    t1 = re.sub(',.([0-9]*).publications', r';\1;', text)
    t2 = re.sub('between.([0-9]*).and.([0-9]*),', r'\1-\2;', t1)
    t3 = re.sub('average.*(...[0-9])', r'\1', t2)
    return t3
```

```python
## the function of this function NLP is to shorten the each Top_Concepts sentence into key info we need
```

```python
def renderdf(textold):
    tokenized_list = re.split('\.',textold)
    list_cleaned = [NLP(text) for text in tokenized_list]
    list_cleaned_df = pd.DataFrame(list_cleaned, index=None, columns=['Overview'])
    list_cleaned_df = list_cleaned_df.iloc[:-1]
    return list_cleaned_df
```

```python
## now to start data restructuring
```

```python
## this renderdf is to generate each Top_Concepts cell of a provider into a single dataframe
```

```python
dfraw["Top_Concepts_Summary"] = dfraw['Top_Concepts'].apply(lambda x: renderdf(x))
```

```python
def getting_end(text):
    ending_year = re.search('-([0-9]+)',text).group(1)
    return ending_year
```

```python
##This function get the last year of publications
```

```python
def range_length(text):
    ending_year = re.search('-([0-9]+)',text).group(1)
    starting_year = re.search('([0-9]+)-',text).group(1)
    range_length = int(ending_year) - int(starting_year)
    if range_length == 0:
        range_length=1
    return range_length
```

```python
##This function get the range_length of publication 
```

```python
dfcleaned["Range_length"] = dfcleaned['Range'].apply(lambda x: range_length(x))
```

```python
dfcleaned["Last_Pub_year"] = dfcleaned['Range'].apply(lambda x: getting_end(x))
```
